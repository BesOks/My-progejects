### task 3
**Пользователи из каких стран совершили наибольшее число успешных бронирований? Укажите топ-5.**

```python
country_top5 = pd.DataFrame(
    bookings.query('is_canceled == 0')
    .country.value_counts()[:5]
    .reset_index()
    .rename(columns={'index':'country', 'country':'bookings'})
)  # interval for top5 contry, sort defolt in descending order
```

```python
country_top5
```

```python
country_top5_bookings = country_top5.country.to_list()
print(*country_top5_bookings, sep=', ')
```

**Answer:** наибольшее число успешных бронирований в PRT, GBR, FRA, ESP, DEU (топ 5)

### task 4
**На сколько ночей в среднем бронируют отели разных типов?**

```python
mean_hotel = (
    bookings.groupby("hotel", as_index=False)
    .agg({"stays_total_nights" : "mean"})
    .rename(columns={'stays_total_nights' : 'mean_night'})
    .round()
)

mean_hotel
```

**Answer:** в отеле City Hotel в среднем бронируют 3 ночи, в отеле Resort Hotel в среднем бронируют 4 ночи

### task 6
**Проанализируйте даты запланированного прибытия. На какой месяц чаще всего успешно оформляли бронь в 2016? Изменился ли самый популярный месяц в 2017?**

```python
arrival_date_2016 = pd.DataFrame(
    bookings.query("arrival_date_year == 2016")
    .value_counts('arrival_date_month')
    .reset_index()
    .rename(columns={0:'bookings'})
)  # booking in 2016 year every month

arrival_date_2016
```

```python
max_arrival_2016 = arrival_date_2016.bookings.max()  # month with maximum arrival in 20216
arrival_date_2016.query('bookings == @max_arrival_2016')
```

```python
arrival_date_2017 = pd.DataFrame(
    bookings.query("arrival_date_year == 2017")
    .value_counts('arrival_date_month')
    .reset_index()
    .rename(columns={0:'bookings'})
)  # booking in 2016 year every month

arrival_date_2017
```

```python
max_arrival_2017 = arrival_date_2017.bookings.max()  # month with maximum arrival in 2017
arrival_date_2017.query('bookings == @max_arrival_2017')
```

**Answer:** в 2016 самый популярный месяц - Октябрь, а в 2017 - Май. 

### task 7
**Сгруппируйте данные по годам и проверьте, на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в каждый из периодов**

```python
canceled_top = pd.DataFrame(
    bookings.query("is_canceled == 1 and hotel == 'City Hotel'") 
    .groupby('arrival_date_year') 
    .arrival_date_month.value_counts()
)

canceled_top
```

```python
max_arrival_2015 = canceled_top.query('arrival_date_year == 2015').arrival_date_month.max()
canceled_top.query('arrival_date_month == @max_arrival_2015')
```

```python
max_arrival_2016 = canceled_top.query('arrival_date_year == 2016').arrival_date_month.max()
canceled_top.query('arrival_date_month == @max_arrival_2016')
```

```python
max_arrival_2017 = canceled_top.query('arrival_date_year == 2017').arrival_date_month.max()
canceled_top.query('arrival_date_month == @max_arrival_2017')
```

**Answer:** самые частые отмены бронирования в отеле City Hotel:  
в 2015 - в Сентябре (1543 раза);  
в 2016 - в Октябре (1947 раз);  
в 2017 - в Мае (2217 раз)

### task 8

### task 9
**Создайте колонку total_kids, объединив children и babies. Для отелей какого типа среднее значение переменной оказалось наибольшим?**

```python
bookings['total_kids']= bookings.children + bookings.babies
```

```python
bookings.groupby('hotel').agg({'total_kids' : 'mean'}).round(2)
```

**Answer:** наибольшее среднее значение у отеля Resort Hotel 

### task 10
**Создайте переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), в противном случае – False. Посчитайте отношение количества ушедших пользователей к общему количеству клиентов, выраженное в процентах (churn rate). Укажите, среди какой группы показатель выше.  
В качестве ответа укажите наибольший % оттока, округленный до 2 знаков после точки (то есть доля 0.24563 будет 24.56% и в ответ пойдёт 24.56)**

```python
bookings['has_kids'] = bookings.total_kids > 0  # true if client have kids
```

```python
bookings
```

```python
bookings.query('is_canceled == 1 and has_kids == False').shape[0]  # count people who canceled the booking and not have kids
```

```python
bookings.query('is_canceled == 1 and has_kids == True').shape[0]  # count people who canceled the booking and have kids
```

```python
# Churn rate who canceled booking (how dependence on kids?)
no_kids_churn = bookings.query('is_canceled == 1 and has_kids == False').shape[0] / bookings.query('has_kids == False').shape[0]

no_kids_churn = round(no_kids_churn * 100, 2)  #precent 
no_kids_churn 
```

**Answer:** Показатель количества ушедших пользователей выше у тех, кто не имеет детей (37,22%)
